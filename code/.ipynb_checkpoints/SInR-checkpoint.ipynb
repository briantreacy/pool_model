```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

class SInR:
    
    def __init__(self, beta, N, t_I, n):
        
        self.beta = beta
        self.N = N
        self.gamma = 1 / t_I
        self.n = n
        
    def fun(self, t, y, beta, N, t_I, n):
        
        gamma = 1 / t_I
        
        S = y[0]
        I = y[1 : n + 1]
        R = y[-1]
        
        d_S_dt = - beta * S * sum(I)
        d_I1_dt = beta * S * sum(I) - n * gamma * I[0]
        d_Ii_dt = [n * gamma * (I[i] - I[i + 1]) for i in range(0, n-1)]
        d_R_dt = n * gamma * I[-1]
        
        dydt = [d_S_dt, d_I1_dt] + d_Ii_dt + [d_R_dt]
        
        return dydt
    
    def sol(self, beta, N, t_I, 
            t_linspace, y0 = None, method = 'LSODA'):
        
        # This generates a solution over a linear time space t_linspace
        
        y0 = [N-1, 1] + [0 for i in range (0, n)]
        
        t0, tf = t_linspace[0], t_linspace[-1]
        
        return solve_ivp(self.fun, (t0, tf), y0, method, t_eval = t_linspace, args = (beta, N, t_I, n))
    
    def traj(self, compartment_string, y):
        
        n = len(y) - 2
        
        if compartment_string == 'S':
            
            return y[0]
        
        elif compartment_string == 'R':
            
            return y[-1]
        
        else:
            
            i = int(compartment_string[1:])
            
            if i < 1 or i > n:
                
                return None
            
            else:
                return y[i]
        
```

```python
R0 = 3
N = 8000000
t_I = 3
t_linspace = np.linspace(0, 33, 10 * 100)
n = 4

beta = R0 / (N * t_I)
gamma = 1 / t_I

SI3R_0 = SInR(beta, N, t_I, n)
sol = SI3R_0.sol(beta, N, t_I, t_linspace).y

```

```python
# Determining r

q = lambda r, n, gamma: 1 / (1 + r / (n * gamma))

L_g = lambda r, n, gamma: (1 / n) * q(r, n, gamma) / (1 - q(r, n, gamma)) * (1 - q(r, n, gamma) ** (n + 1))

L_g_dr = lambda r, n, gamma: (- 1 / n) * 1 / (r + n * gamma) * q(r, n, gamma) / ((1 - q(r, n, gamma) ** 2)) * (1 - (n + 1) * q(r, n, gamma) ** n + n * q(r, n, gamma) ** (n + 1))

f = lambda r, n, gamma, beta, N: L_g(r, n, gamma) - beta * N / gamma
f_dr = lambda r, n, gamma: L_g_dr(r, n, gamma)

```

```python
# Visual check of the trajectories

fig, axes = plt.subplots(2, 2, gridspec_kw={'height_ratios': [1, 2]})

# Plotting of S and R trajectories

axes[0, 0].plot(t_linspace, SI3R_0.traj("S", sol))
axes[0, 0].set_title("S")
axes[0, 1].plot(t_linspace, SI3R_0.traj("R", sol))
axes[0, 1].set_title("R")

for i in range(1, n + 1):
    
    Ii_string = "I" + str(i)
    
    axes[1, 0].plot(t_linspace, SI3R_0.traj(Ii_string, sol), label = Ii_string)
    axes[1, 0].legend()
    axes[1, 0].set_title(r"$I_k$")

fig.delaxes(axes[1, 1])

# Plotting of the 
plt.tight_layout()
plt.show()
```

```python

proportions = [SI3R_0.traj("I" + str(i), sol) / sum(sol[1:-1]) for i in range(1, n + 1)]


fig, axes = plt.subplots()

for i, proportion in enumerate(proportions):
    
    axes[i].plot(t_linspace, proportion, label = "I" + str(i + 1))

plt.tight_layout()
plt.show()
    
    
```
